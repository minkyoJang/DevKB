```python
elicehotel_8 = "EliceHotel_8.txt"
elicehotel_9 = "EliceHotel_9.txt"

def txtToDict(filename):
    userDict = {}
    with open(filename) as file:
        for line in file:
            a = line.find(':')
            
            key = line[:a]
            data = line[a+1:]
            
            userDict[key] = data.strip()
            # 함수를 완성하세요.
            
            
        return userDict

# txtToDict 함수를 활용하여 txt파일을 딕셔너리로 저장합니다.
eliceHotel8 = txtToDict(elicehotel_8)
eliceHotel9 = txtToDict(elicehotel_9)

# eliceHotel8과 eliceHotel9에서 Values만 추출하여 
# 각각 eliceHotel8_value와 eliceHotel9_value에 저장합니다.
# 그리고 set형식으로 변환합니다
eliceHotel8_value = set(eliceHotel8.values())
eliceHotel9_value = set(eliceHotel9.values())


# 교집합을 출력합니다.
print(eliceHotel8_value & eliceHotel9_value)
```

```python
# 답
import csv

price = [1000,1200,800,900,1000,500,600,1200,1500,2000]

def convertDict(csvfile, price):

    Dict = {}
    
    with open(csvfile) as f:
        csv_data = csv.reader(f)
        li = []
        for data in csv_data:
            li.extend(data)   
    
    for i,v in enumerate(li): 
        Dict[v] = price[i]
     
    return Dict
    

machine = convertDict('item.csv',price)
print(machine['orange-juice'])
```

```python
import csv
import json

def get_popular_speaking(n) :
    talks = []
    with open('ted.csv', encoding = 'utf-8') as f:
        reader = csv.reader(f)
        for row in reader :
            if int(row[16]) >= n :
                talks.append((row[14], int(row[16])))
        Final = sorted(talks, key = lambda x : -x[1])
    return Final

def main():
    n = int(input())
    print(get_popular_speaking(n))


if __name__ == "__main__":
    main()
```
